## 1. Import libraries and load the dataset

In this section, the required libraries are imported and the CSV file is loaded into a pandas DataFrame.


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Improve chart style
sns.set_theme(style="whitegrid")

# Load the dataset
df = pd.read_csv("university_student_data.csv")

# Display the first rows
df.head()
```

## 2. Initial exploratory analysis

The dataset contains information for the years 2015 to 2024 and includes two academic terms per year: Spring and Fall.


```
# Basic information about the dataset
print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isnull().sum())

df.describe()
```

## 3. Description of the variables

| Column | Meaning |
|---|---|
| Year | Academic year of the record |
| Term | Academic term: Spring or Fall |
| Applications | Number of student applications received |
| Admitted | Number of students admitted |
| Enrolled | Number of students who enrolled |
| Retention Rate (%) | Percentage of students who continued in the institution |
| Student Satisfaction (%) | Percentage score that reflects student satisfaction |
| Engineering Enrolled | Students enrolled in Engineering |
| Business Enrolled | Students enrolled in Business |
| Arts Enrolled | Students enrolled in Arts |
| Science Enrolled | Students enrolled in Science |


## 4. Retention rate trends over time

The following graph shows the average retention rate by year.


```
retention_by_year = df.groupby("Year")["Retention Rate (%)"].mean()

plt.figure(figsize=(10, 5))
retention_by_year.plot(marker="o")
plt.title("Retention Rate Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Retention Rate (%)")
plt.ylim(80, 95)
plt.grid(True)
plt.show()
```

## 5. Student satisfaction scores by year

This visualization compares average student satisfaction across years.


```
satisfaction_by_year = df.groupby("Year")["Student Satisfaction (%)"].mean()

plt.figure(figsize=(10, 5))
satisfaction_by_year.plot(kind="bar")
plt.title("Student Satisfaction by Year")
plt.xlabel("Year")
plt.ylabel("Student Satisfaction (%)")
plt.ylim(70, 95)
plt.show()
```

## 6. Comparison between Spring and Fall terms

The dataset contains the same values for Spring and Fall in each year. Therefore, this comparison is expected to show no difference between the two terms.


```
term_comparison = df.groupby("Term")[["Applications", "Admitted", "Enrolled"]].sum()

term_comparison.plot(kind="bar", figsize=(10, 5))
plt.title("Comparison Between Spring and Fall Terms")
plt.xlabel("Term")
plt.ylabel("Total Students")
plt.xticks(rotation=0)
plt.show()

term_comparison
```

## 7. Additional trend visualization with seaborn

The following chart uses `sns.lineplot()` to compare retention and satisfaction trends over time.


```
trend_df = df.groupby("Year", as_index=False)[["Retention Rate (%)", "Student Satisfaction (%)"]].mean()
trend_long = trend_df.melt(
    id_vars="Year",
    value_vars=["Retention Rate (%)", "Student Satisfaction (%)"],
    var_name="Indicator",
    value_name="Percentage"
)

plt.figure(figsize=(10, 5))
sns.lineplot(data=trend_long, x="Year", y="Percentage", hue="Indicator", marker="o")
plt.title("Retention and Satisfaction Trends")
plt.ylabel("Percentage")
plt.show()
```

## 8. Main findings

1. Retention increased from **85% in 2015** to **90% in 2024**, although there was a decrease in 2020.
2. Student satisfaction increased from **78% in 2015** to **88% in 2024**, with a small decline in 2020.
3. Applications increased from **2,500** to **3,500** during the observed period.
4. Enrolled students increased from **600** to **800**.
5. Spring and Fall terms show the same values in the dataset, so no differences are observed between them.
6. Engineering has the highest number of enrolled students among the departments.

## 9. Actionable insight

The university should investigate the factors that caused the decline in retention and satisfaction in 2020 and strengthen the strategies that produced the positive recovery from 2021 onward. In addition, departments with lower enrollment, especially Science, could benefit from targeted promotion and student support actions.
